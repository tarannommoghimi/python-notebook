**Question one**

```python
data["DOB"]=pd.to_datetime(data["DOB"])
today = pd.Timestamp.today()
data["age"]=((today- data["DOB"]).dt.days/365).astype(int)
data[["DOB","age"]]
```

```python
data["group by age"]=pd.cut(
    data["age"],
    bins=[0, 25, 40,60,150],
    labels=["<25","25-40","40-60","60<"],
    right=False
)
data[["age","group by age"]],
```

```python
data["group by age"].value_counts()
```

```python
avg_absence=(
    data.groupby("Position")["Absences"].mean().astype(int)
    .sort_values(ascending=False)
)
print(avg_absence)
print("______________________________________")
```

```python
avg_salary=(
    data.groupby("Position")["Salary"].mean()
    .sort_values(ascending=False)
)
print(avg_salary)
```

**Question three**

```python
terminate=data["DateofTermination"].notna()
percentage=terminate.mean()*100
print(f"percentage of employees who left: {percentage:.2f}%")
```

```python
leftcompany = data[data["DateofTermination"].notna()]
single = (leftcompany["MaritalDesc"]=="Single").sum()
married = (leftcompany["MaritalDesc"] == "Married").sum()

print(f"Singles who left the company: {single}")
print(f"Marrieds who turnedover: {married}")
```

Based on the resuls, yes, single employees left the company more often than married employees. This suggests that employee turnover was higher among single employees.we should notice that, this result is based on the number of employees who left and doesnt necessarily mean that being single is the cause of higher turnover.

```python
import pandas as pd

data["DateofHire"] = pd.to_datetime(data["DateofHire"])
data["DateofTermination"] = pd.to_datetime(data["DateofTermination"])

leftdf = data[data["DateofTermination"].notna()]
result = leftdf["DateofTermination"] - leftdf["DateofHire"]
#per year
average_years = result.dt.days.mean() / 365
print(f"Average years of employees who work before they left: {average_years:.2f} years")

#per day
#average_days = result.dt.days.mean()
#print(f"Average day of employees who work befor they left: {average_days:.2f} days")
```

The average tenure of employees who left the company was 3.46 years. This means that employees stayed with the company for about 4 years on average before leaving.
