**Question three**
finding outliar with IQR

```python
import numpy as np
sales=[12,15,14,10,13,50,14,16,15,14,13,12,14,15,11,13,14,13,12,200]
med=np.median(sales)
print("median=",med)
q1=np.percentile(sales, 25)
print("q1=",q1)
q3=np.percentile(sales, 75)
print("q3=",q3)
IQR=q3-q1
print("IQR=",IQR)
lower=q1 - 1.5 * IQR
upper=q3+1.5*IQR
outliar=[x for x in sales if x<lower or x>upper]
print("lower=",lower)
print("upper=:",upper)
print("outlier=", outliar)

```

**three-sigma method**

```python
mean = np.mean(sales)
std = np.std(sales)

lower = mean - 3 * std
upper = mean + 3 * std

outliers = [x for x in sales if x < lower or x > upper]

print("mean =", mean)
print("std =", std)
print("lower=", lower)
print("Upper=", upper)
print("outliers =", outliers)
```

**comparsion between IQR and 3-sgma method**  

IQR method identifies outliers based on first and third quartlies, so its less affected br extreme values. its suiable for skewed data.
3-sigma methos identifies using the mean and std. since both are influencedby extreme values, this method is more sensitive to outliers, and it work best for normally distributed data.
in this dataset the IQR methode detected 50 and 200 as outliers,while the other one only give us one(200).
based on this matter IQR is more effective for this dataset.

```python
#based on IQR, deleting 200,50
import numpy as np

data = [12,15,14,10,13,50,14,16,15,14,13,12,14,15,11,13,14,13,12,200]

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
IQR = q3 - q1

lower = q1 - 1.5 * IQR
upper = q3 + 1.5 * IQR

clean_data = [x for x in data if lower <= x <= upper]

print("clean Data =", clean_data)
print("new Mean =", np.mean(clean_data))
print("new Median =", np.median(clean_data))
```

**yes**  
the mean changed a lot after removing the outliers data,while thee median has a slightly change.
